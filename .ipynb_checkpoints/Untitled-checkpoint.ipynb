```python
import pandas as pd
import numpy as np
```

```python
confirmed = pd.read_csv('time_series_covid19_confirmed_global.csv')
confirmed = confirmed.fillna('')
deaths = pd.read_csv('time_series_covid19_deaths_global.csv')
deaths = deaths.fillna('')
recovered = pd.read_csv('time_series_covid19_recovered_global.csv')
recovered = recovered.fillna('')
```

```python
columns = confirmed.columns.values
```

```python
col_country = columns[:4]
```

```python
col_dates = columns[4:]
```

```python
dates_confirmed = confirmed[col_dates]
dates_deaths = deaths[col_dates]
dates_recovered = recovered[col_dates]
states_countries_data = confirmed[col_country]
```

```python
import mysql.connector
```

```python
def createDB():
    print("open connection")
    conn = mysql.connector.connect(
        host = "localhost",
        user = "root",
        password = ""
    )

    cursor = conn.cursor()
    
    print("Create database")
    #create database
    cursor.execute('CREATE DATABASE IF NOT EXISTS COVID_DB')
    
    print("Create tables")
    #create tables
    tables = [
        'CREATE TABLE IF NOT EXISTS COVID_DB.country_states (id INT PRIMARY KEY AUTO_INCREMENT, country VARCHAR(25), state VARCHAR(25) NULL, lat DECIMAL(10,8),lon DECIMAL(11,8))',
        'CREATE TABLE IF NOT EXISTS COVID_DB.dates (id INT PRIMARY KEY AUTO_INCREMENT, date_info DATE, confirmed INT DEFAULT 0, death INT DEFAULT 0, recovered INT DEFAULT 0, country_state_id INT, CONSTRAINT FK_country_dates FOREIGN KEY (id) REFERENCES country_states(id))'
    ]
    
    #create all tables
    for t in tables:
        cursor.execute(t)
    cursor.close()
    conn.close()
    print("connection closed")
```

```python
createDB()
```

```python
conn = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "",
    database = "COVID_DB"
)

cursor = conn.cursor()

for i,s in enumerate(np.array(states_countries_data)):
    country = s[1]
    state = s[0]
    lat = s[2]
    lon = s[3]
    
    sql = "INSERT INTO country_states (country,state,lat,lon) VALUES (%s, %s, %s, %s)"
    val = (country,state,lat,lon)
    cursor.execute(sql, val)
    country_state_id = cursor.lastrowid
    for d in dates_confirmed:
        n_confirmed = -1 #if is -1 not found data for that date
        n_death = -1
        n_recovered = -1
        try:
            n_confirmed = int(dates_confirmed.loc[(confirmed['Country/Region'] == country) & (confirmed['Province/State'] == state)][d][0])
            n_death = int(dates_confirmed.loc[(deaths['Country/Region'] == country) & (deaths['Province/State'] == state)][d][0])
            n_recovered = int(dates_confirmed.loc[(recovered['Country/Region'] == country) & (recovered['Province/State'] == state)][d][0])
        except:
            pass
        add_detail = ("INSERT INTO dates "
              "(date_info, confirmed, death, recovered, country_state_id) "
              "VALUES (%(date_info)s, %(confirmed)s, %(death)s, %(recovered)s, %(country_state_id)s)")
        
        data_detail = {
            'date_info': d,
            'confirmed': n_confirmed,
            'death': n_death,
            'recovered': n_recovered,
            'country_state_id': country_state_id
        }
        
        print(data_detail)
        cursor.execute(add_detail, data_detail)

# Make sure data is committed to the database
conn.commit()
        
        #sql = "INSERT INTO dates (date_info, confirmed, death, recovered, country_state_id) VALUES (%s,%s,%s,%s,%s)"
        #val = (d,n_confirmed,n_death,n_recovered,country_state_id)
        #cursor.execute(sql, val)
```
